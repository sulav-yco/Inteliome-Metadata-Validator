```python
def save_schemas_to_yaml(schema_dict, output_path: str):
        serializable_dict = {
            table_name: schema.model_dump(mode="python")  # Converts Pydantic model to plain dict
            for table_name, schema in schema_dict.items()
        }
        
        with open(output_path, "w") as f:
            yaml.dump(serializable_dict, f, sort_keys=False, default_flow_style=False)
```

```python
save_schemas_to_yaml(schema_dict = schema,output_path = "./generator/output/schema/schema.yml")
```
